```python
import os
import cv2
from PIL import Image
import numpy as np
import pickle


BASE_DIR = 'E:/AI class/DSP_project/'
image_dir=os.path.join(BASE_DIR,"images") #locate image

face_cascade = cv2.CascadeClassifier('cascade/data/haarcascade_frontalface_alt2.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()


current_id=0
label_ids={}
y_labels=[]
x_train=[]
#see image
for root,dirs,files in os.walk(image_dir):
    for file in files:
        if file.endswith("png") or file.endswith("jpg"):
            path=os.path.join(root,file)
            label=os.path.basename(os.path.dirname(path)).replace(" ","-").upper()
            #print(label,path)
            if not label in label_ids:
                label_ids[label]=current_id
                current_id+=1
            id_=label_ids[label]
            #print(label_id)
            #y_labels.append(label) #some number
            #x_path.append(path) #verify that image,turn into numpy arra,gray
            #pil is python image library
            pil_image=Image.open(path).convert("L") #convert into greyscale
            size=(550,550)
            final_image=pil_image.resize(size,Image.ANTIALIAS)
            image_array=np.array(final_image,"uint8") #Unsigned integer (0 to 255)
            #print(image_array)
            #dectection
            faces=face_cascade.detectMultiScale(image_array,scaleFactor=1.5,minNeighbors=5)
            for (x,y,w,h) in faces:
                #ROI Region Of Interest
                roi=image_array[y:y+h,x:x+w] #(ystart yend same for x)
                x_train.append(roi)
                y_labels.append(id_)
#print(x_train)
#print(y_labels)
with open("labels.pickle","wb") as f: #wb as writing bytes
    pickle.dump(label_ids,f) #save label ids in files
    
recognizer.train(x_train,np.array(y_labels))
recognizer.save("trainner.yml") #YAML is a human-readable data-serialization language. It is commonly used for configuration files
```

```python
import numpy as np
import cv2
import pickle

face_cascade = cv2.CascadeClassifier('cascade/data/haarcascade_frontalface_alt2.xml')
recognizer = cv2.face.LBPHFaceRecognizer_create()
recognizer.read("trainner.yml") #train data
labels={"person":1}

with open("labels.pickle","rb") as f:   #wb as reading bytes
    og_labels=pickle.load(f)    #load label ids in files
    labels={v:k for k,v in og_labels.items()}   #invert the labels
cap=cv2.VideoCapture(0)

while True:
    #capture fame by frame
    ret,frame=cap.read()
    #convert frame into grey for faical recogitation
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    
    faces=face_cascade.detectMultiScale(gray,scaleFactor=1.5,minNeighbors=5)
    for (x,y,w,h) in faces:
        #print(x,y,w,h)
        #ROI Region Of Interest
        roi_gray=gray[y:y+h,x:x+w] #(ystart yend same for x)
        roi_color=frame[y:y+h,x:x+w]
        #recogniser deep learnig model
        id_,conf = recognizer.predict(roi_gray) #predict region of interest
        if conf>=45: # and conf<=85:
            print(id_)
            print(labels[id_])
            font= cv2.FONT_HERSHEY_SIMPLEX
            name =labels[id_]
        
        img_item='7.png'
        cv2.imwrite(img_item,roi_gray)
        
        #draw a rectangle arround face
        color=(255,0,0) #BGR 0-255 
        stroke=2 #thickness of rectangle
        end_cord_x=x+w
        end_cord_y=y+h
        cv2.rectangle(frame,(x,y),(end_cord_x,end_cord_y),color,stroke)
        color=(255,255,255)
        stroke=2
        cv2.putText(frame,name,(x,y),font ,1,color,stroke,cv2.LINE_AA)
        
        
    #display the resulting image
    cv2.imshow('Frame',frame)
    if cv2.waitKey(20) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
